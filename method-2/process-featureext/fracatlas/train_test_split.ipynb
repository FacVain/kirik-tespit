```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
FractAtlas = pd.read_csv('../../../data/FracAtlas/dataset.csv')
FractAtlas.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(FractAtlas.drop('fractured', axis=1), FractAtlas['fractured'], test_size=0.2, random_state=42, stratify=FractAtlas['fractured'])
```

```python
import pickle as pkl

#to save it
with open("./test-train-dataset/train.pkl", "wb") as train_file:
    pkl.dump([X_train, y_train], train_file)

with open("./test-train-dataset/test.pkl", "wb") as test_file:
    pkl.dump([X_test, y_test], test_file)
```
