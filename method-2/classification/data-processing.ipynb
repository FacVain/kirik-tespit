```python
import pickle as pkl
import pandas as pd
```

```python
with open("../process-featureext/test-train-dataset/train.pkl", "rb") as train_file:
    [X_train, y_train] = pkl.load(train_file)

with open("../process-featureext/test-train-dataset/test.pkl", "rb") as test_file:
    [X_test, y_test] = pkl.load(test_file)
```

```python
train = pd.DataFrame(X_train)[["image_id", "hand", "leg", "hip", "shoulder", "mixed"]].rename(columns={"image_id": "image_file"}).assign(label=y_train)
train
```

```python
canny_hough = pd.read_csv("../process-featureext/canny_hough_features.csv")
sobel_harris = pd.read_csv("../process-featureext/sobel_harris_features.csv")
```

```python
train = train.merge(canny_hough, on="image_file").merge(sobel_harris, on="image_file")
train
```

```python
train.to_csv("train.csv", index=False)
```

```python
test = pd.DataFrame(X_test)[["image_id", "hand", "leg", "hip", "shoulder", "mixed"]].rename(columns={"image_id": "image_file"}).assign(label=y_test)
test = test.merge(canny_hough, on="image_file").merge(sobel_harris, on="image_file")
test.to_csv("test.csv", index=False)
```
